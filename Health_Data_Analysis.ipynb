AN ANALYSIS OF HEALTHCARE DATASET FOCUSING ON DISEASE, TREATMENT TYPE, HOSPITALS, AND RECOVERY RATE.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#The data set
from google.colab import files
uploaded=files.upload()
file_name = next(iter(uploaded))
df = pd.read_excel(file_name)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
#CHECKING FOR MISSING VALUES
df.isnull()
```

```python
#DROPPING COLUMN WITH MISSING VALUES-(the cause of Death Column)
df.dropna(inplace=True)
df
```

```python
#CHECK FOR NUMBER OF DISEASES AND TREATMENT MODES undertaken
df.groupby('Disease')['Treatment'].size()
```

```python
#Visual of the hospital Disease and Treatment Modes undertaken
treatment_count=df.groupby('Disease')['Treatment'].size()
plt.figure(figsize=(8,4))
treatment_count.plot(kind='barh',color='blue')
plt.title('Treatments and number of Times handled')
plt.xlabel('Treatment times')
plt.ylabel('Disease')
plt.tight_layout()
plt.show()
```

Malaria is the Disease which was handled most number of Times. Tuberculosis and Asthma followed second. Dengue was the least Disease handled.

```python
#CHECK FOR GENDER POPULARITY PER HOSPITALS
df.groupby('Gender')['Hospital'].value_counts()
```

```python
df_grouped = df.groupby(['Gender', 'Hospital']).size().reset_index(name='Count')

plt.figure(figsize=(8, 6))
sns.barplot(data=df_grouped, x='Gender', y='Count', hue='Hospital')

plt.title('Hospital popularity on genders')
plt.xlabel('Gender')
plt.ylabel('Hospital')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

Lifeline Hospital is the Most prefered Hospital among the two genders. City Hospital is the most unpopular among the female patients while Sunrise Clinic is the least prefered Hospital among the male patients.

```python
#CHECK FOR HOSPITALS AND NUMBER OF TREATMENTS HANDLED
df.groupby('Hospital')['Treatment'].size()
```

```python
#Visual of the hospital distribution and patient handled
treatment_count=df.groupby('Hospital')['Treatment'].size()
plt.figure(figsize=(8,4))
treatment_count.plot(kind='barh',color='green')
plt.title('Hospitals and Number of Treatments handled')
plt.xlabel('Hospital')
plt.ylabel('Treatment')
plt.tight_layout()
plt.show()
```

Lifeline Clinic handled most Treatments per patients while Sunrise Clinic handled the least amount of Treatments.

```python
#DISEASE DISTRIBUTION PER GENDER COUNT
df.groupby('Gender')['Disease'].size()
```

```python
#checking for disease that impacted specific gender most
df.groupby('Gender')['Disease'].value_counts()
```

```python
# VISUAL FOR DISEASE DISTRIBUTION PER GENDER
df_grouped = df.groupby(['Gender', 'Disease']).size().reset_index(name='Count')

plt.figure(figsize=(8, 6))
sns.barplot(data=df_grouped, x='Gender', y='Count', hue='Disease')

plt.title('Disease Distribution per gender')
plt.xlabel('Gender')
plt.ylabel('Count of Disease')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

Malaria is the most prevalent disease among the two genders. Covid-19 is the least recorded disease among the females while Dengue was the least recorded disease among the males.

```python
#CHECK for count of disease and treatment status
df.groupby('Disease')['Treatment Status'].value_counts()
```

```python
#CHECKING FOR THE STATUS OF THE DISEASES BASED TREATMENT STATUS
grouped = df.groupby('Disease')['Treatment Status'].value_counts()
df_counts=grouped.unstack(fill_value=0)
highest_recovered = df_counts['Recovered'].idxmax()
highest_deceased = df_counts['Deceased'].idxmax()
highest_under_treatment = df_counts['Under Treatment'].idxmax()
print("Disease with highest recovery Rate:", highest_recovered)
print("Disease with highest deceased Rate:", highest_deceased)
print("Disease with highest under treatment:", highest_under_treatment)
```

From the data set, Malaria as a Disease has the highest recovery rate while at the same time has the highest number of patients still under treatment. Asthma is the Disease that has the highest mortality rate.

```python
#CHECKING FOR THE OUTCOME OF TREATMENT BASED ON TREATMENT STATUS
df.groupby('Treatment')['Treatment Status'].value_counts()
```

```python
#CHECKING FOR THE STATUS OF TREATMENT STATUS BASED ON TREATMENT PROVIDED
groupedT=df.groupby('Treatment')['Treatment Status'].value_counts()
df_counts=groupedT.unstack(fill_value=0)
highest_recovered = df_counts['Recovered'].idxmax()
highest_deceased = df_counts['Deceased'].idxmax()
highest_under_treatment = df_counts['Under Treatment'].idxmax()
print("Treatment with highest recovery Rate:", highest_recovered)
print("Treatment with highest deceased Rate:", highest_deceased)
print("Treatment with highest under treatment:", highest_under_treatment)
```

From the data set, Dialysis as a form of Treatment has the highest recovery rate, at the same time it had the highest mortality rate and also has the highest patients under treatment.


```python
#CHECKING FOR TREATMENT WITH THE RATE OF ADMISSION DAYS
df.groupby('Treatment')['Days Admitted'].value_counts()
```

```python
# CHECKING FOR TREATMENT WITH THE TOTAL NUMBER OF ADMISSION DAYS
treatment_admission_days = df.groupby('Treatment')['Days Admitted'].sum()

highest_treatment_days = treatment_admission_days.idxmax()
lowest_treatment_days = treatment_admission_days.idxmin()

print("Treatment with the highest total days of admission:", highest_treatment_days)
print("Treatment with the lowest total days of admission:", lowest_treatment_days)
```

Dialysis Treatment was found to have the highest days of admission when administering Treatment. On the other hand, Medication was found to have the less number of days of admission.

```python
#CHECKING FOR TREATMENT OPTIONS WITH AGE
df.groupby('Treatment')['Age (Complete Years)'].value_counts()
```

```python
#Visualized Treatment Options with Age
df_grouped = df.groupby(['Treatment', 'Age (Complete Years)']).size().reset_index(name='Count')

plt.figure(figsize=(8, 6))
sns.barplot(data=df_grouped, x='Treatment', y='Count', hue='Age (Complete Years)')

plt.title('Treatment Distribution on age ')
plt.xlabel('Treatment')
plt.ylabel('Age')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

Dialysis as a mode of Treatment is more prevalent with an increase in age of patients. secondly, many aged patients also received treatment in surgery, and physical therapy. Chemotherapy, Ventilation support, Radiotherapy and Medication apears to not attract an older age group. Many younger patients appeared to take medications as the form of Treatment.
